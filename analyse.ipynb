```python
import numpy as np #mathematical tools
import matplotlib.pyplot as plt #plot charts
import pandas as pd #
from sklearn.linear_model import LinearRegression
#print(type(dataset))
import statsmodels.api as sm

    

```

```python
dataset = pd.read_csv('analyse.csv')#input dataset

Y = dataset.population#[:] means all values
X = dataset[["speed","volume"]]
Y1 = dataset["18_25"]#[:] means all values
X1 = dataset[["Dine-in","Beverages","Desserts","Fast Food","Bakery","Cafe"]]
#print(X1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.4, random_state=101)
X1_train, X1_test, Y1_train, Y1_test = train_test_split(X1, Y1, test_size=0.4, random_state=101)
#dataset.reshape(-1, 1)
lm = LinearRegression()
lm1 = LinearRegression()
lm.fit(X_train,Y_train)
lm1.fit(X1_train,Y1_train)
predictions = lm.predict(X_test)
predictions1 = lm1.predict(X1_test)
plt.scatter(Y_test,predictions)
plt.ylabel("dependent variable")
plt.xlabel("independent variables")
plt.show()
plt.scatter(Y1_test,predictions1,color="red")
plt.ylabel("dependent variable")
plt.xlabel("independent variables")
plt.show()
```

```python
X_=sm.add_constant(X)
results=sm.OLS(Y,X_).fit()
results.params
reg_eq=results.params[1]*X+results.params[2]*X+results.params[0]
plt.plot(X,reg_eq)
```

```python
X_1=sm.add_constant(X1)
results1=sm.OLS(Y1,X_1).fit()
results1.params
reg_eq1=results1.params[1]*X1+results1.params[2]*X1+results1.params[0]
plt.plot(X1,reg_eq1)
```
